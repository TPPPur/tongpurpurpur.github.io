# (a) Explain in words how graphs such as the Pen’s Parade, Lorenz curve and TIP curve are constructed. How is the Generalised Lorenz Curve (GLC) related to the Lorenz curve?

## Pen's parade: 
* data: using individual data that is proportional to one's income
* order: rank from the poorest to the richest 
* y-axis is income level, x-axis is the cumulative population share
* we can interpret a point on the x-axis as the one's income ranking among the sample, expressed in percentage

## Lorenz curve:
* raw data: same as Pen's
* data manipulation: measured in cumulative income shares
* order: same as Pen's 
* y-axis is the cumulative income share, x-axis is the same as Pen's 
* so Lorenz curve has a range of [0:1] in both dimensions 
* I think lorenz curve is useful because a 45 degree line can be drawn representing equlity line, so a Lorenz curve of a distibution will show how unequal the distribution is, in terms of income

## TIP curve:
* raw data: individual income
* data manipulation: to construct a normalised poverty gap for people who's income is below a poverty line (an assumed value) , poverty value minus one's income and then divided by the poverty line; for the subsample who's income is equal or higher than the poverty line, we record them as zero
* y-axis is cumulative normalised poverty gap per capita, x-axis is cumulative population share
* from TIP we can see 3 things:
  * how poor poor people can be on averate
  * what's the propotion of poor people in the distribution
  * distribution of poor people, are they just below the poverty line or in absolute poverty

## GLC and LC
* GLC is LC streched up at each point by the cumulative average income at that point 
* GLC can not only show the level of inequlity, but can also indicate how large the pie is
